```python
import csv
import numpy as np
from matplotlib import pyplot as plt
import os
import fnmatch
from time import sleep

from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier as KNN

%matplotlib inline
```

```python
def parse_horizon(infile):
    ils, xls, values = ([],[],[])
    with open(infile, 'r') as f:
        reader=csv.reader(f, delimiter='\t')
        for il, xl, value in reader:
            ils.append(int(il)-1)
            xls.append(int(xl)-1)
            point = float(value)
            values.append(point)
    
    value_array = np.array(values)
    inlines = np.array(ils)
    xlines = np.array(xls)
    horizon = np.zeros((np.amax(inlines)-np.amin(inlines) +1, 
                         np.amax(xlines) - np.amin(xlines)+1))
    horizon[inlines-np.amin(inlines), xlines - np.amin(xlines)] = values
    
    return horizon
```

```python
horA = parse_horizon('HorA.txt')
horB = parse_horizon('HorB.txt')
horC = parse_horizon('HorC.txt')
horD = parse_horizon('HorD.txt')
horE = parse_horizon('HorE.txt')

print np.amin(horB)
```
